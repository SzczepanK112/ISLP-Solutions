### 2. It was stated in the text that classifying an observation to the class for which equation (4.17) is largest is equivalent to classifying an observation to the class for which equation (4.18) is largest. Prove that this is the case. In other words, under the assumption that the observations in the $k$-th class are drawn from a $\mathcal{N}(\mu_k, \sigma^2)$ distribution, the Bayes classifier assigns an observation to the class for which the discriminant function is maximized.

The equation (4.17) is given as:

$$
p_k(x) = \frac{\pi_k \sqrt{\frac{1}{2\pi\sigma}} \exp\left(-\frac{1}{2\sigma^2} (x - \mu_k)^2\right)}{\sum_{l=1}^K \pi_l \sqrt{\frac{1}{2\pi\sigma}} \exp\left(-\frac{1}{2\sigma^2} (x - \mu_l)^2\right)} \tag{4.17}
$$

The equation (4.18) is given as:

$$
\delta_k(x) = \frac{x \cdot \mu_k}{\sigma^2} - \frac{\mu_k^2}{2\sigma^2} + \log(\pi_k) \tag{4.18}
$$

<br>

$$
\max (p_k(x)) = \max\left(\pi_k\exp\left(-\frac{1}{2\sigma^2} (x - \mu_k)^2\right)\right)  \newline
$$

<br>
We can take natural logarithm from the part of the equation above because it preserves the order of maximization  

<br>
<br>

$$
log\left(\pi_k\exp\left(-\frac{1}{2\sigma^2} (x - \mu_k)^2\right)\right) 
$$

<br>

$$
log(\pi_k) -\frac{1}{2\sigma^2} (x - \mu_k)^2 = log(\pi_k) - \frac{1}{2\sigma^2}(x^2 - 2x\mu_k + {\mu_k}^2) = 
log(\pi_k) - \frac{x^2}{2\sigma^2} + \frac{2x\mu_k}{2\sigma^2} - \frac{{\mu_k}^2}{2\sigma^2}
$$

<br>

Because the term $ -\frac{x^2}{2\sigma^2} $ is constant with respect to $k$ we can ommit it in the maximizaiton. Finally we get:

<br>

$$
\frac{x \cdot \mu_k}{\sigma^2} - \frac{\mu_k^2}{2\sigma^2} + \log(\pi_k) = \delta_k(x)
$$

### 4. When the number of features p is large, there tends to be a deterioration in the performance of KNN and other local approaches that perform prediction using only observations that are near the test observation for which a prediction must be made. This phenomenon is known as the curse of dimensionality, and it ties into the fact that curse of dimensionality non-parametric approaches often perform poorly when p is large. We will now investigate this curse.  

(a) Suppose that we have a set of observations, each with measurements on p = 1 feature, X. We assume that X is uniformly(evenly)distributed on [0, 1]. Associated with each observation is a response value. Suppose that we wish to predict a test observation’s responseusing only observations that are within 10 % of the range of X closest to that test observation. For instance, in order to predict theresponse for a test observation with X = 0.6, we will use observations in the range [0.55, 0.65]. On average, what fraction of the available observations will we use to make the prediction?  

- Since $X$ is uniformly distributed on [0, 1], the probability of any observation falling within this range is the fraction of the total range that this interval occupies. The total length of the range of $X$ is 1, so:

    $ 0.1 (10\%) $

(b) Now suppose that we have a set of observations, each with
measurements on p = 2 features, X1 and X2. We assume that
(X1, X2) are uniformly distributed on [0, 1] × [0, 1]. We wish to
predict a test observation’s response using only observations that
are within 10 % of the range of X1 and within 10 % of the range
of X2 closest to that test observation. For instance, in order to
predict the response for a test observation with X1 = 0.6 and
X2 = 0.35, we will use observations in the range [0.55, 0.65] for
X1 and in the range [0.3, 0.4] for X2. On average, what fraction
of the available observations will we use to make the prediction?  

- $0.1 \times 0.1 = 0.01 (1\%)$

(c) Now suppose that we have a set of observations on p = 100 features. Again the observations are uniformly distributed on each
feature, and again each feature ranges in value from 0 to 1. We
wish to predict a test observation’s response using observations
within the 10 % of each feature’s range that is closest to that test
observation. What fraction of the available observations will we
use to make the prediction?  

- For each feature, the range is 10% of the total range [0, 1], which is 0.1. So, for each of the 100 features, we are considering a 10% slice of the total range. The "region" of observations considered is the volume of a 100-dimensional hypercube with side length 0.1. The volume $ V $ of a hypercube with side length $ s $ in $ p $-dimensional space is:

$$
V = s^p
$$

- In our case, $ s = 0.1 $ and $ p = 100 $, so:

$$
V = 0.1^{100}
$$

(d) Using your answers to parts (a)–(c), argue that a drawback of
KNN when p is large is that there are very few training observations “near” any given test observation.  

- The key insight from parts (a) to (c) is that as the number of features $ p $ increases, the volume of the region around a test observation decreases exponentially.

(e) Now suppose that we wish to make a prediction for a test observation by creating a p-dimensional hypercube centered around
the test observation that contains, on average, 10 % of the training observations. For p = 1, 2, and 100, what is the length of each
side of the hypercube? Comment on your answer.
Note: A hypercube is a generalization of a cube to an arbitrary
number of dimensions. When p = 1, a hypercube is simply a line
segment, when p = 2 it is a square, and when p = 100 it is a
100-dimensional cube

- The fraction of training observations contained within the hypercube is equal to the volume of the hypercube. If we want the hypercube to contain 10% of the training observations, we set the volume to 0.1:

$$
\text{Volume of hypercube} = s^p = 0.1
$$

  - To find the length \( s \) of the hypercube in terms of \( p \), we solve:

$$
s^p = 0.1
$$

  - Taking the \( p \)-th root:

$$
s = 0.1^{1/p}
$$

For specific values of \( p \):

- For \( p = 1 \):
  $$
  s = 0.1^{1/1} = 0.1
  $$

- For \( p = 2 \):
  $$
  s = 0.1^{1/2} \approx 0.316
  $$

- For \( p = 100 \):
  $$
  s = 0.1^{1/100} \approx 0.977
  $$


As the number of dimensions $ p $ increases, the side length of the hypercube approaches 1, which indicates that the hypercube encompasses a larger portion of the feature space. In very high-dimensional spaces, the side length is very close to 1, meaning that nearly all observations are contained in the hypercube.

### 6. Suppose we collect data for a group of students in a statistics classwith variables X1 = hours studied, X2 = undergrad GPA, and Y= receive an A. We fit a logistic regression and produce estimated coefcient, $\beta_{0} = −6$, $\beta_{1} = 0.05$, $\beta_{2} = 1$.
(a) Estimate the probability that a student who studies for 40 h and
has an undergrad GPA of 3.5 gets an A in the class.

$$
P(Y=1) = \frac{\exp(\beta_0 + \beta_1 X_1 + \beta_2 X_2)}{1 + \exp(\beta_0 + \beta_1 X_1 + \beta_2 X_2)} = 
\frac{\exp(-6 + 0.05 \times 40 + 1 \times 3.5)}{1 + \exp(-6 + 0.05 \times 40 + 1 \times 3.5)} \approx 0.38
$$

(b) How many hours would the student in part (a) need to study to
have a 50 % chance of getting an A in the class?

A 50% chance of getting an A corresponds to $ P(Y=1) = 0.5 $  
At $ P(Y=1) = 0.5 $ the logistic function satisfies:
$$
\frac{\exp(z)}{1 + \exp(z)} = 0.5 \quad \Rightarrow \quad \exp(z) = 1
$$

This implies that:
$$
z = \beta_0 + \beta_1 X_1 + \beta_2 X_2 = 0
$$

Rearranging for $ X_1 $ (hours studied):
$$
X_1 = \frac{-\beta_0 - \beta_2 X_2}{\beta_1} = \frac{6 - 1 \times 3.5}{0.05} = 50
$$

$$

$$

$$

$$
P(X = 4) = \left( \frac{1}{\sqrt{72\pi}} e^{-\frac{1}{2}} \cdot 0.8 \right) + \left( \frac{1}{\sqrt{72\pi}} e^{-\frac{2}{9}} \cdot 0.2 \right) =

```python
# Now dropping all predictors with p-value > 0.05

X_train_small = X_train.drop(columns=['zn', 'tax', 'medv', 'ptratio', 'rm', 'chas', 'age'])
X_test_small = X_test.drop(columns=['zn', 'tax', 'medv', 'ptratio', 'rm', 'chas', 'age'])

glm_small = sm.GLM(y_train, X_train_small, family=sm.families.Binomial())
glm_small_results = glm_small.fit()

summarize(glm_small_results)
```

```python
# Now dropping some predictors

X_train_mid = X_train.drop(columns=['chas', 'zn', 'rm', 'tax'])
X_test_mid = X_test.drop(columns=['chas', 'zn', 'rm', 'tax'])

glm_mid = sm.GLM(y_train, X_train_mid, family=sm.families.Binomial())
glm_mid_results = glm_mid.fit()

summarize(glm_mid_results)
```

```python
y_pred = (glm_mid_results.predict(X_test_mid) > 0.5).astype(int)
confusion_table(y_pred, y_test)
```

```python
accuracy_score(y_test, y_pred)  # Best accuracy achieved for all predictors
```

```python
# LDA

train_sets = [X_train, X_train_small, X_train_mid]
test_sets = [X_test, X_test_small, X_test_mid]
labels = ['all predictors', 'predictors with viable p-value for linear regression', 'some predictors chosen manually']


lda = LDA(store_covariance=True)

for i in range(3):

  lda.fit(train_sets[i], y_train)

  y_pred_lda = lda.predict(test_sets[i])


  print(f'LDA data for {labels[i]}: \n'
        'Confusion table:\n'
        f'{confusion_table(y_test, y_pred_lda)}\n'
        f'Accuracy score: {accuracy_score(y_pred_lda, y_test)}\n\n')

# Best accuracy achieved for smallest predictor dataset,
# this dataset also achieved smallest ammount of false positives but biggest ammount of false negatives
```

```python
# Naive Bayes

nb = GaussianNB()

for i in range(3):

  nb.fit(train_sets[i], y_train)

  y_pred_nb = nb.predict(test_sets[i])
  ct = confusion_table(y_test, y_pred_nb)

  recall = ct.loc[True, True] / (ct.loc[True, True] + ct.loc[True, False])
  precision = ct.loc[True, True] / (ct.loc[True, True] + ct.loc[False, True])

  print(f'Naive Bayes output for {labels[i]}: \n\n'
        'Confusion table:\n'
        f'{ct}\n\n'
        f'Accuracy score: {accuracy_score(y_pred_nb, y_test)}\n'
        f'Recall score: {recall}\n'
        f'Precision score: {precision}\n\n')

# Naive bayes displays worse performance than Linear Discriminant Analysis,
# highest ammount of cases where crime ratio is above median were identified by the model with all predictors
```

```python
# KNN

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled_small = scaler.fit_transform(X_train_small)
X_test_scaled_small = scaler.fit_transform(X_test_small)

X_train_scaled_mid = scaler.fit_transform(X_train_mid)
X_test_scaled_mid = scaler.fit_transform(X_test_mid)

train_sets_scaled = [X_train_scaled, X_train_scaled_small, X_train_scaled_mid]
test_sets_scaled = [X_test_scaled, X_test_scaled_small, X_test_scaled_mid]

knn_params_per_set = []

for i in range(3):
  best_k = None
  best_pred = None
  best_accuracy = 0
  
  for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_sets_scaled[i], y_train)
    y_pred_knn = knn.predict(test_sets_scaled[i])
    acc = accuracy_score(y_test, y_pred_knn)

    if acc > best_accuracy:
      best_k = k
      best_pred = y_pred_knn
      best_accuracy = acc

  # knn_params_per_set.appen((k, acc))

  ct = confusion_table(y_test, best_pred)

  recall = ct.loc[True, True] / (ct.loc[True, True] + ct.loc[True, False])
  precision = ct.loc[True, True] / (ct.loc[True, True] + ct.loc[False, True])

  print(f'KNN output for {labels[i]} with best K estimated to be {best_k}: \n\n'
        'Confusion table:\n'
        f'{ct}\n\n'
        f'Accuracy score: {best_accuracy}\n'
        f'Recall score: {recall}\n'
        f'Precision score: {precision}\n\n')
  

# KNN for predictors with p-value below 0.05 achieves accuracy score of 95% and precision of 98.5%, pretty good
```
