```python
Auto = load_data("Auto")
y = Auto['mpg'] # mpg -> miles per galon
X = MS(['horsepower']).fit_transform(Auto)
model = sm.OLS(y, X)

# Slight negative relationship between horsepower and miles per galon
results = model.fit()
summarize(results)
```

```python

ax = Auto.plot.scatter('horsepower', 'mpg')
xlim = ax.get_xlim()
ylim = [results.params[1] * xlim[0] + results.params[0], results.params[1] * xlim[1] + results.params[0]]
ax.grid(True)
ax.plot(xlim, ylim, 'r--', linewidth=3)
```

```python
ax = plt.subplots(figsize=(8,8))[1]
ax.scatter(results.fittedvalues , results.resid)
ax.set_xlabel('Fitted value')
ax.set_ylabel('Residual')
ax.axhline(0, c='k', ls='--')
```

```python
infl = results.get_influence()
ax = plt.subplots(figsize=(8,8))[1]
ax.scatter(np.arange(X.shape[0]), infl.hat_matrix_diag)
ax.set_xlabel('Index')
ax.set_ylabel('Leverage')
```

```python
plt.figure(figsize=(8, 6))
sns.residplot(x=results.fittedvalues, y=results.resid)
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
plt.title("Residuals vs Fitted")
plt.show()
```

```python
interaction_terms = ['cylinders:displacement', 'horsepower:displacement', 'horsepower:acceleration', 'origin:year']

model = smf.ols(formula='mpg ~ ' + ' + '.join(terms) + ' + ' + ' + '.join(interaction_terms), data=Auto)
results = model.fit()
summarize(results)
```

```python
log_terms = ['np.log(' + term + ')' for term in terms]
sqrt_terms = ['np.sqrt(' + term + ')' for term in terms]
pow_terms = ['np.power(' + term + ', 2)' for term in terms]


model = smf.ols(formula='mpg ~ ' + ' + '.join(log_terms), data=Auto)
results = model.fit()
summarize(results)

# For logarithmized terms cilinders and displacement are not related to response according to statistical methods
```

```python
model = smf.ols(formula='mpg ~ ' + ' + '.join(sqrt_terms), data=Auto)
results = model.fit()
summarize(results)

# Cylinder count, displacement and acceleration (in square root form) does not have a notable impact on mpg
```

```python
model = smf.ols(formula='mpg ~ ' + ' + '.join(pow_terms), data=Auto)
results = model.fit()
summarize(results)

# In model with squared predictors an intercept value is not statistically significant which is disturbing
```

```python
model = smf.ols(formula='mpg ~ ' + ' + '.join(log_terms) + ' + ' + ' + '.join(terms), data=Auto)
results = model.fit()
summarize(results)

# If we create a model composed of linear terms and logarithmized terms, the standardized error increases for linear terms
```
